```python
import pandas as pd
import numpy as np
```

```python
# the data is taken from the WDI data base https://datacatalog.worldbank.org/dataset/world-development-indicators
# for the speed purposes the data fis already extracted for Russia for the period 1994-2017
df = pd.read_excel('RUS.xlsx')
df.head()
```

```python
# extract the necessary variables and transpose the dataframe
df = df[df['Indicator Code'].isin(['NY.GDP.MKTP.CN', 'NE.CON.PRVT.CN', 'NE.EXP.GNFS.CN', 'NE.IMP.GNFS.CN', 'NE.GDI.TOTL.CN', 'NE.CON.GOVT.CN', 'BN.CAB.XOKA.GD.ZS', 'NY.GDP.DEFL.ZS', 'SP.POP.TOTL'])] 
df = df.drop(['Indicator Code'], axis=1)
df = df.T 
df.columns = df.iloc[0]
df = df[df['Current account balance (% of GDP)'] != 'Current account balance (% of GDP)']
df = df.apply(pd.to_numeric)
df.head()
```

```python
# save the data
df.to_excel("data.xlsx")
```
